```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
dir_geoparsed_results = os.path.dirname(os.getcwd())
```

## Reading Toponym Resolution Results of Edinburgh Geoparser

```python
df_lgl_edinburgh = pd.read_csv(dir_geoparsed_results+'\\geoparsed-results\\lgl_geocoded_results_Edinburgh_Geoparser.csv')
df_lgl_edinburgh.head()
```

```python
sns.displot(df_lgl_edinburgh, x="median_error_distance",stat="probability",bins=20)
plt.xlabel('MdnED (km)', fontsize=14)
plt.ylabel('Normalized Frequency', fontsize=14)
plt.title('LGL - Edinburgh Geoparser', fontsize = 18)
plt.ylim(0,1)
plt.xlim(0,17500)
```
