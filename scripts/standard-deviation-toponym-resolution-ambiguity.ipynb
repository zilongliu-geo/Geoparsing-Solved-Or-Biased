```python
import pandas as pd
import os
```

```python
dir_geoparsed_results = os.path.dirname(os.getcwd())
```

## Calculate Standard Deviation of MdnED (Produced by Edinburgh Geoparser) for Top 10 Highly Ambiguous Toponym in WikToR

```python
df_wiktor_result = pd.read_csv(dir_geoparsed_results+'\\geoparsed-results\\wiktor_geocoded_results_Edinburgh_Geoparser.csv')
del df_wiktor_result['Unnamed: 0']
df_wiktor_result.head()
```

```python
df_wiktor_result.groupby(['name']).size().sort_values(ascending=False).head(20)
```

```python
for name in df_wiktor_result.groupby(['name']).size().sort_values(ascending=False).head(10).index:
    print(name, df_wiktor_result[['name','median_error_distance']].groupby(['name']).get_group(name).agg({'median_error_distance':['std']}))
```
